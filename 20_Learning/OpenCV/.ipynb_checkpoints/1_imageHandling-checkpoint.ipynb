# Image Handling

- Image: 2D grid of pixels(Matrix\Tables) or Matrix of numbers
    - Every Image is like Excel Sheet which contain number of rows and columns which is known as pixels
- Pixel: Our Smallest Unit Of Pixel - If picture is Colored. It basically Contain RGB Values in range of 0-225
- width: Total Number of pixels horizontally(Left to Right)
- length: Total Number of pixels Vertically(Top to Bottom)
- Color Channels: A Channel which basically stored 1 part of color information in each pixel
    - Geyscale 1 Channel
    - RGB : 3 Channels
            - [255,0,0] RED, [0,255,0] Green, [0,0,255]: Blue
    - BGR : OpenCV Formet
