# Edge Detection and thresholding

- Canny Edge Detection is an algorithm that finds sharp changes (edges) in brightness or intensity within an image.

## Why its used

- To identify boundaries of shapes, letters, or objects.

    - Used before tasks like:

    - OCR (detect text areas)

    - Image segmentation

    - Object detection

    - Contour detection

### Apply Canny Edge Detection(v imp)
- edges = cv2.Canny(image, threshold1, threshold2)
- threshold1: Lower limit for edge detection(weak edge)
- threshold2: Upper limit (strong edge)
