# Image Filtering and Bluring

- Image filtering means modifying each pixels value based on the values of its neighboring pixels.
- Blurring means reducing image noise and detail by averaging nearby pixel values.
    - type of image filtering
- **NOISE** is unwanted random variations in pixel values.
      - Imagine you scan a handwritten note.
      - The text is dark and important 
      - But the paper might have small dots, shadows, or ink smudges 
      - These tiny unwanted marks are noise.
- Smoothing = reducing or averaging pixel intensity differences between neighboring pixels to make the image more uniform.
    - Basically, smoothing = blurring out the noise.
    - It makes edges and background look cleaner.
-A kernel is a small matrix (like 33 or 55) that slides over the image to modify each pixel.
    - Its also called a filter or convolution matrix.

# Gaussian Blur

- Gaussian Blur is a type of image smoothing that uses the Gaussian function to blur an image
- it reduces noise and detail smoothly while keeping edges relatively sharp

- Imagine youre cleaning an image with a soft brush 
- the center (brush tip) cleans strongly, and it fades softly toward the sides.
- Thats exactly how a Gaussian kernel works.

- Syantax
- **blurred_image = cv2.GaussianBlur(image, (ksize_X,ksize_Y), sigmaX)**

- Ksize should always odd , it should never be even
- sigmaX: Controls how much blur to apply along X-axis(if 0 than OpenCV decides by itself how much blur it need to do)

```python
import cv2
image = cv2.imread(r"G:\Project\PDF_TO_TEXT\1_pdf_to_image\output_images\TOC\page_1.jpeg")
blurred = cv2.GaussianBlur(image, (7,7),0)
cv2.imshow("Original image",image)
cv2.imshow("Gaussian blurred image",blurred)
cv2.waitKey(0)
cv2.destroyAllWindows
```

- more the size of Kernel more the blur happended
